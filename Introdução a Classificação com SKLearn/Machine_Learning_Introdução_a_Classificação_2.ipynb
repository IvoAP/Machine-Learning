```python
import pandas as pd
```

```python
uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
```

#Ler o CSV

```python
dados = pd.read_csv(uri)
dados.head()
```

#Renomear colunas 

```python
mapa = {
    'home' : 'principal',
    'how_it_works' : 'como_funciona',
    'contact' : 'contato',
    'bought' : 'comprou'

}

dados = dados.rename(columns=mapa)
dados.head()
```

# Pegar o x (features) dos nossos dados


```python
x = dados[['principal','como_funciona', 'contato']]
x.head()
```

#Pegar o nosso y (resultado)

```python
y = dados[['comprou']]
y.head()
```

```python
dados.shape
```

# Sabendo que temos 99 linhas e 4 colunas, iremos pegar para trenar até 75 do x usar os dados restantes para x

```python
treino_x = x[:75]
treino_y = y[:75]
teste_x = x[75:]
teste_y = y[75:]
```

```python
print("Usaremos %d de dados para treino e %d de dados para teste" % (len(treino_x), len(treino_y)))
```

```python
print(f"Usaremos {len(treino_x)} de dados para treino e {len(teste_x)} de dados para teste")
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)
acuracia = accuracy_score(teste_y, previsoes) * 100 # taxa de acerto
```

```python
print("A minha acurácia foi de  %.2f%%" % (acuracia))
```

#Usando a biblioteca que separa treino e teste

```python
from sklearn.model_selection import train_test_split
SEED = 20

treino_x , teste_x, treino_y, teste_y = train_test_split(x,y, random_state=SEED,test_size=0.25)
print(treino_x.shape)
print(teste_x.shape)
```

```python
print("Usaremos %d de dados para treino e %d de dados para teste" % (len(treino_x), len(treino_y)))
```

```python
modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)
acuracia = accuracy_score(teste_y, previsoes) * 100 # taxa de acerto

```

```python
print("A minha acurácia foi de  %.2f%%" % (acuracia))
```

```python
treino_y.value_counts()
```

```python
teste_y.value_counts()
```

#Como vimos os nossos dados existem mais pessoas que não compraram no teste e e pouquissimas compraram no teste. Precisamos treinar os dados e garantir uma igual proporção, colocando o parametro stratify no nosso train_test_split

```python
treino_x , teste_x, treino_y, teste_y = train_test_split(x,y, 
                                                         random_state=SEED,
                                                         test_size=0.25,
                                                         stratify= y)
print(treino_x.shape)
print(teste_x.shape)

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)
acuracia = accuracy_score(teste_y, previsoes) * 100 # taxa de acerto

```

```python
print("A minha acurácia foi de  %.2f%%" % (acuracia))
```
