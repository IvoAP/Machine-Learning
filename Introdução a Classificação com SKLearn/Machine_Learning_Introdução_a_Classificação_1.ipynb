```python
#features (1 -> yes e  0 -> no)
# long hair 
# short leg
# faz auau

pig1 = [0, 1, 0]
pig2 = [0, 1, 1]
pig3 = [1, 1, 0]

dog1 = [0, 1, 1]
dog2 = [1, 0, 1]
dog3 = [1, 1, 1]

#f(x) = y
train_x = [pig1, pig2, pig3, dog1, dog2, dog3]

# my classes: pig = 1 and dog = 0

train_y = [1,1,1,0,0,0] #labels
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(train_x,train_y)
```

```python
mysterious_animal = [1,1,1]

# predict espera um array 2D, não somente um item
model.predict([mysterious_animal])
```

```python
mysterio1 = [1, 1, 1]
mysterio2 = [1, 1, 0]
mysterio3 = [0, 1, 1]

test_x = [mysterio1, mysterio2, mysterio3]

predictions = model.predict(test_x)

print (predictions)
```

```python
# Imagine que nós já sabiamos o resultado e o mesmo era:
test_y  = [0, 1, 1]
```

```python
print(predictions)
```

```python
print(test_y)
```

```python
# Vamos comparar as predições com os resultados corretos

test_y == predictions
```

```python
# Vamos calcular nossos pontos

hits = (predictions == test_y).sum()
print(hits)
```

```python
# Vamos calcular o tamanho total da amostra usada para testes

total_tests = len(test_x)
print(total_tests)
```

```python
# Vamos ver nosssa tacha de acertor em porcento
hit_rate = (hits/total_tests) * 100
print(f'Hit rate: {hit_rate}')
```

```python
# Como usar a biblioteca metrics do SKLearn metrics para calcular a taxa de acerto
from sklearn.metrics import accuracy_score

hit_rate = accuracy_score(test_y, predictions)
print('Hit rate: %.2f'% (hit_rate*100))
```
